```python
import numpy as np 
import pandas as pd 
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)});
```

```python
df_tiktok = pd.read_csv('social media influencers - tiktok.csv')
df_instagram = pd.read_csv('social media influencers - instagram.csv')
df_youtube = pd.read_csv('social media influencers - youtube.csv')
```

# Instagram

## Top countries audience

```python
df_instagram.head()
```

```python
df_instagram.isnull().sum()
```

```python
df_instagram.dropna(inplace=True)
```

```python
chart = sns.countplot(x="Audience country(mostly)", data=df_instagram, order = df_instagram['Audience country(mostly)'].value_counts().index)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```
